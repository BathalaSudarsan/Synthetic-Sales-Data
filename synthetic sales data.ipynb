```python
import pandas as pd
from faker import Faker
import random
from flask import send_file

# Create an instance of the Faker class
faker = Faker()

# Define the number of synthetic records to generate
num_records = 100  #change the input according to our requirement

# Generate synthetic customer data
customer_data = []
for _ in range(num_records):
    customer = {
        'CustomerID': faker.uuid4(),
        'FirstName': faker.first_name(),
        'LastName': faker.last_name(),
        'Email': faker.email(),
        'Phone': faker.phone_number(),
        'Address': faker.address(),
        'City': faker.city(),
        'Country': faker.country(),
        'BirthDate': faker.date_of_birth(minimum_age=18, maximum_age=80).strftime('%Y-%m-%d'),
    }
    customer_data.append(customer)

# Generate synthetic product data
product_data = []
for _ in range(num_records):
    product = {
        'ProductID': faker.uuid4(),
        'ProductName': faker.word(),
        'Description': faker.text(),
        'Price': round(faker.random_int(min=10, max=1000)),
    }
    product_data.append(product)

# Generate synthetic sales transaction data
sales_data = []
for _ in range(num_records):
    transaction = {
        'TransactionID': faker.uuid4(),
        'CustomerID': random.choice(customer_data)['CustomerID'],
        'ProductID': random.choice(product_data)['ProductID'],
        'Quantity': random.randint(1, 10),
        'SalesDate': faker.date_between(start_date='-1y', end_date='today').strftime('%Y-%m-%d'),
        'Price': random.choice(product_data)['Price'],
    }
    sales_data.append(transaction)

# Create Pandas DataFrames from the synthetic data
df_customers = pd.DataFrame(customer_data)
df_products = pd.DataFrame(product_data)
df_sales = pd.DataFrame(sales_data)


# Concatenate the dataframes
Sales_data = pd.concat([df_customers, df_products, df_sales], axis=1)

# Display the concatenated data
Sales_data.head()

#download and save the csv file where you saved the programming file automatically
Sales_data.to_csv('Sales_data.csv', index=False)  #remove hash symbol before Sales then you get csv file to your local disk


```
